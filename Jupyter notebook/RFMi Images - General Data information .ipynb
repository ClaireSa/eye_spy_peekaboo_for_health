```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('RFMiD_Training_Labels.csv')
```

```python
data.head()
```

```python
num_rows, num_cols = data.shape

print(f"The number of rows is: {num_rows}")
print(f"The number of columns is: {num_cols}")
```

```python
disease_risk_counts = data['Disease_Risk'].value_counts()
```

```python
if 1 in disease_risk_counts:
    print(f"Number of '1's : {disease_risk_counts[1]}")
else:
    print("Number of '1's : 0")

if 0 in disease_risk_counts:
    print(f"Number of '0's : {disease_risk_counts[0]}")
else:
    print("Number of '0's : 0")
```

```python
#Iterate through each column and count occurences of '1' and '0'
for column in data.columns:
    if data[column].dtype == 'int64' or data[column].dtype == 'float64':  
        value_counts = data[column].value_counts()
        ones = value_counts.get(1, 0)  
        zeros = value_counts.get(0, 0)  
        
        print(f"Column '{column}':")
        print(f"Number of '1's: {ones}")
        print(f"Number of '0's: {zeros}")
        print("-----------")
```
